```python
import matplotlib.pyplot as plt

# Setter opp aksen
fig, ax = plt.subplots()
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_xlabel('Kvantum (Q)')
ax.set_ylabel('Pris (P)')
ax.set_title('Monopolprissetting')

# Etterspørselskurven
demand_curve = ax.plot([10, 0], [0, 10], label='Etterspørselskurve')

# Marginalkostnadskurven
mc_curve = ax.plot([0, 10], [2, 2], label='Marginalkostnadskurve')

# Finner monopolpris og kvantum
monopoly_price = 6
monopoly_quantity = 4

# Monopolprisen og kvantum
ax.scatter(monopoly_quantity, monopoly_price, color='red')
ax.annotate('Pris', xy=(monopoly_quantity, monopoly_price), xytext=(4.5, 6), 
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)
ax.annotate('Kvantum', xy=(monopoly_quantity, monopoly_price), xytext=(1.5, 4), 
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)

```
