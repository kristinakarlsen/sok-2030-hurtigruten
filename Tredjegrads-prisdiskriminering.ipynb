Hvordan hurtgruten kan øke sin profitt ved tredjegrads prisdiskriminering

```python
#importerer de nødvendige pakkene
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt

#definerer etterspørselsfunksjonen
def demand(Q):
    return 9000-250*Q

#definerer marginal inntekts funksjonen
def marginalrevenue(Q):
    return 9000-500*Q

#definerer marginal kostnadsfunksjonen
def marginalcost(Q):
    return 500
```

```python
Q=sp.symbols('Q', real=True, positive=True)

#lager ligning som setter marginal inntekt lik marginal kostnad, for å kunne finne produksjonsnivået
equ=sp.Eq(marginalrevenue(Q),marginalcost(Q))
equ
```

```python
#løser ligningen og runder av svaret, finner optimalt kvantum
Q_equ=sp.solve(equ,Q)[0]
Q_equ
round(Q_equ,2)
```

```python
#løser ligningen og runder av svaret, finner optimalt kvantum
Q_equ=sp.solve(equ,Q)[0]
Q_equ
round(Q_equ,2)
```

```python
#setter det optimale kvantum inn i etterspørselsfunksjonen for å finne optimal pris
p_eq=demand(Q_equ)
p_eq
```

```python
#finner salget for den ordinære kundegruppen
def Q(P):
    return 20-P/500 #definerer den inverse etterspørselen for ordinær gruppe
print(Q(4750))#løser for Q (antall solgte billetter) når prisen er 4750 (uniform pris)
```

```python
#finner kvantum salg for student kundegruppen 
def Q(P):
    return 16-P/500 #definerer den inverse etterspørselen for student
print(Q(4750))#løser for Q når prisen er 4750
```

```python
# Marginalinntekt ordinær 
def MR(Q):
    return 10000-1000*Q
print(MR(10.5))
```

```python
# Marginalinntekt student 
def MR(Q):
    return 8000-1000*Q
print(MR(6.5))
```

```python
# Finner hurtigrutens profitt ved uniform pris
def profitt(Q):
    return (p_eq-marginalcost(Q))*Q_equ
print (profitt(Q_equ))
```

```python
Q = np.linspace(0, 60, 10) #definerer et intervall for hvilke verdier som inkluderes i beregningen 

#legger til restriksjoner på y-aksen, for hvilke tall som fremheves
plt.ylim(0,11000)
plt.xlim(0,40)

#lager variabler med etterspørsel, marginal inntekt og marginal kostnad
demand_values = demand(Q) 
marginalrevenue_values = marginalrevenue(Q)
marginalcost_values = np.full_like(Q, marginalcost(Q)) #holder marginal kostnaden konstant

#genererer plot som viser etterspørsel, marginalkostnad og marginal inntekt
plt.plot(Q, demand_values, color = "yellow", label="Etterspørsel")
plt.plot(Q, marginalrevenue_values, color = "green", label="Marginal Inntekt")
plt.plot(Q, marginalcost_values, color = "pink", label="Marginal Kostnad")

likevekt_Q = 17 #lager verdien for likevektskavantum som en variabel
likevekt_P = 4750 #lager verdien for likevektspris som en variabel

#lager linjer som hviser likevekts pris og likevekts kvantum
plt.hlines(likevekt_P, 0,likevekt_Q, color='gray', linestyle='--', label="likevekts pris")
plt.vlines(likevekt_Q, 0, likevekt_P, color='gray', linestyle='--', label="likevekts kvantum")


#inkluderer likvekts kvantum og likevekts pris på x-aksen og y-aksen
xticks = list(plt.gca().get_xticks()) #beholder de opprinnelige verdiene på x-aksen
if likevekt_Q not in xticks:
    xticks.append(likevekt_Q) #legger til en spesifik verd å fremheve på x-asken
xticks = sorted([tick for tick in xticks if tick >= 0]) #sorterer listen med x-verdier og lager listen til et set
plt.xticks(xticks) #legger til listen i plottet

yticks = list(plt.gca().get_yticks())
if likevekt_P not in yticks:
    yticks.append(likevekt_P)
yticks = sorted([tick for tick in yticks if tick >= 0])
plt.yticks(yticks)

#gir "legend"-en en gitt plasering
plt.legend(loc='upper right')

#gir navn på aksene
plt.ylabel('Pris')
plt.xlabel('Antall biletter')

#gir plottet en overskrift
plt.title("uniform prising")

#hviser plottet
plt.show()
```
