Tilpassning unifor pris

```python
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt

def demand(Q):
    return 9000-250*Q

def marginalrevenue(Q):
    return 9000-500*Q

def marginalcost(Q):
    return 500
```

```python
Q=sp.symbols('Q', real=True, positive=True)
equ=sp.Eq(marginalrevenue(Q),marginalcost(Q))
equ
```

Optimail tilpassning uniform prising

```python
Q_equ=sp.solve(equ,Q)[0]
Q_equ
round(Q_equ,2)
```

```python
p_eq=demand(Q_equ)
p_eq
```

```python
#salg ordinær
def Q(P):
    return 20-P/500
print(Q(4750))
```

```python
#salg studenter
def Q(P):
    return 16-P/500
print(Q(4750))
```

```python
# Marginalinntekt ordinær P = 10 000 - 1000Qo
def MR(Q):
    return 10000-1000*Q
print(MR(10.5))
```

```python
# Marginalinntekt student P = 8000 - 1000Qs
def MR(Q):
    return 8000-1000*Q
print(MR(6.5))
```

```python
# Profitt
def profitt(Q):
    return (p_eq-marginalcost(Q))*Q_equ
print (profitt(Q_equ))
```

```python
Q = np.linspace(0, 60, 10)

demand_values = demand(Q)
marginalrevenue_values = marginalrevenue(Q)
marginalcost_values = np.full_like(Q, marginalcost(Q))

plt.plot(Q, demand_values, color = "yellow", label="Etterspørsel")
plt.plot(Q, marginalrevenue_values, color = "green", label="Marginal Inntekt")
plt.plot(Q, marginalcost_values, color = "pink", label="Marginal Kostnad")

likevekt_Q = 17
likevekt_P = 4750

plt.hlines(likevekt_P, 0,likevekt_Q, color='gray', linestyle='--', label="likevekts pris")
plt.vlines(likevekt_Q, 0, likevekt_P, color='gray', linestyle='--', label="likevekts kvantum")

plt.legend(loc='upper right')

plt.ylabel('Pris')
plt.xlabel('Antall biletter')

plt.title("uniform prising")

plt.ylim((-2,10000))

plt.show()
```
